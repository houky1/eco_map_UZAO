```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Akademicheskaya_station.CSV')
df = df.iloc[3:]
Q1 = df['PM2.5'].quantile(0.25)
Q3 = df['PM2.5'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['PM2.5'] >= lower_bound) & (df['PM2.5'] <= upper_bound)]
df
```

```python
mean = df['PM2.5'].mean()
median = df['PM2.5'].median()

print(mean)
print(median)
```
