```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
df = pd.read_csv('Konkovo_park.CSV')

df.drop(columns = ['Time_sec'], inplace=True)
df
```

```python
plt.plot(df.index, df['Humidity'], marker = 'o', linestyle = '-', color = 'r')

plt.title('График влажности')
plt.xlabel('Порядковый номер')
plt.ylabel('Влажность')

plt.show()
```

Теперь удалим первые пять значений

```python
df.drop(df.index[:5], inplace=True)
df
```

```python
plt.plot(df.index, df['Humidity'], marker = 'o', linestyle = '-', color = 'r')

plt.title('График влажности')
plt.xlabel('Порядковый номер')
plt.ylabel('Влажность')

plt.show()
```

```python
Q1 = df['Humidity'].quantile(0.25)
Q3 = df['Humidity'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['Humidity'] >= lower_bound) & (df['Humidity'] <= upper_bound)] 
df
```

```python
plt.plot(df.index, df['Humidity'], marker = 'o', linestyle = '-', color = 'r')

plt.title('График влажности')
plt.xlabel('Порядковый номер')
plt.ylabel('Влажность')

plt.show()
```
