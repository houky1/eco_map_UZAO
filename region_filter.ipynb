```python
import json

# 1. Открываем файл и загружаем данные
with open("moscow_hoods.json", "r", encoding="utf-8") as f_in:
    raw_data = json.load(f_in)

target_district = "ЮЗАО"

# 2. Фильтруем список словарей напрямую
filtered_features = [
    item for item in raw_data 
    if isinstance(item, dict) and item.get("district") == target_district
]

# 3. Преобразуем отфильтрованный список в стандартный GeoJSON
new_geojson = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            # Здесь можно добавить координаты (geometry), если они понадобятся для карты
            "geometry": {
                "type": "Polygon",
                "coordinates": [] 
            },
            "properties": {
                "hood": item.get("hood"),
                "district": item.get("district"),
                "stations": item.get("stations")
            }
        }
        for item in filtered_features
    ]
}

# 4. Сохраняем в новый файл
with open("Filtered_South_West.json", "w", encoding="utf-8") as f_out:
    json.dump(new_geojson, f_out, ensure_ascii=False, indent=4)

print(f"Сохранено районов для ЮЗАО: {len(filtered_features)}")
```
